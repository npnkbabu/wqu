```python
def plotYields(start="2001-01-01",    end="2001-12-31",    resampleDuration="M"):
    dt_yields1={series_id:get_yield_data(series_id,start=start,end=end) for series_id in series_ids}
    yields1 = pd.DataFrame(dt_yields1)
    yields1.columns  =  ['1 Month', '3 Month', '6 Month', '1 Year', '2 Year', '3 Year', '5 Year', \
                      '7 Year', '10 Year', '20 Year', '30 Year']
    yields1.sort_index(inplace=True)
    small_tenure = yields1.dropna(axis=1, how="all")
    small_tenure.index = small_tenure.index.map(pd.to_datetime)
    small_tenure = small_tenure.resample(resampleDuration).mean()
    small_tenure.index = small_tenure.index.strftime("%Y-%b")
    data = small_tenure.T
    plt.figure(figsize=(15,5))
    plt.plot(data)
    plt.legend(data.columns,loc="best")
    plt.show()
```

```python
# 1981-1982 high inflation and interest rates
# what happened : because of high inflation, govt increases interest rates.
# effect on producer : production is reduced as borrowing money became costly.
# effect on consumer : consumption is constly becoz of high prices (inflation) at the same time saving is better option becoz of high interest rates.
# Overall effect on yields : Since interest rates are more definetly existing bond prices (short term) will go down, so yield became high
plotYields(start="1981-01-01",end="1982-12-31",resampleDuration="Y")
```

```python
# 2001 dot com bubble created inflation so interest rates are lowered to simulate economy
# what happened : recession
# govt action on interest rates : lowered to simulate economy, so yeild should go down.
# short term yiled going down faster than long term yileld which tells its bull steepening
plotYields(start="2001-01-01",end="2001-12-31",resampleDuration="Q")
```

```python
# 2007-2009 crash.
# what happened : inflation followed by recession
# govt action on interest rates : 2007 interest rate decreased from 5.25% to 2%.
# short term yiled going down faster than long term yileld which tells its bull steepening
plotYields(start="2007-01-01",end="2007-12-31",resampleDuration="M")
```

```python
# 2008-2009 crash.
# what happened : inflation followed by recession
# govt action on interest rates : continue to lower the rates from 2008 to 2016
# long term yields increase faster than short term yields, its bear steepening
plotYields(start="2008-01-01",end="2016-12-31",resampleDuration="Y")
```

```python
# 2017 to 2018 
# what happened : inflation followed by recession
# govt action on interest rates : 2017 to 2019 gradually increased interest rates
# we see a gradual increase of yields
plotYields(start="2017-01-01",end="2018-12-31",resampleDuration="Y")
```

```python
# 2019 - 2022 Covid
# What happened: recession
# govt action on interest rates: decreased to stimulate the economy, so yield should go down.
# If we take 2019 as reference then from 2020-2021 short term yield decreased and went almost 0, its bull steepening. 
# 2022: From 2022 Govt started increasing interest rates so we see short term bonds yields started increasing. bear flattening
plotYields(start="2019-01-01",end="2022-12-31",resampleDuration="Y")
```

```python
# 2022 and 2023 govt started increasing interest rates
# 2023 and 2024: short term bond yield increased faster than long term. bear flattening
# 2025:  short term yields gone down so again bull steepening. 
# looks like bull steepening and bear flattening is a cycle
plotYields(start="2023-01-01",end="2025-12-31",resampleDuration="Y")
```
